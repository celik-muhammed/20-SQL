- SQLite Aggregiate

```python
q = """
select  origin
        , avg(mpg) avg_mpg
        , avg(cylinders) avg_cylinders
        , avg(horsepower) avg_horsepower
        , avg(weight) avg_weight
        , avg(acceleration) avg_acceleration
from    df
group by origin;        
"""
sqldf(q)
```

```python
q = """
select  *
from    df a, df b
WHERE   a.name = b.name;
"""
sqldf(q).head()
```

```python
q = """
select  *
from    df
where   name in (select name 
                 from df 
                 where mpg > 25);
"""
sqldf(q)
```

```python
q = """
select  *
from    df a
where   EXISTS (select  1 
                from    df b  
                where   mpg > 25 
                        and a.name = b.name);
"""
sqldf(q)
```
